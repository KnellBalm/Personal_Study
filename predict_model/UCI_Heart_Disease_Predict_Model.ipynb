```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
 
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 
  
# metadata 
print(heart_disease.metadata) 
  
# variable information 
print(heart_disease.variables) 
df = pd.concat([X, y], axis=1)
    
```

## Target ì—­í•  ì»¬ëŸ¼ ì •ì˜
- ì‹¬ìž¥ë³‘ ìœ ë¬´ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì´ì§„ ê°’ ì»¬ëŸ¼ì´ ìžˆëŠ” ì§€ í™•ì¸
- num ì»¬ëŸ¼ì€ 0 ì´ë©´ ì •ìƒ, 1~4ëŠ” ì§ˆí™˜ì´ ìžˆê¸° ë•Œë¬¸ì— boolean ê°’ìœ¼ë¡œ ë³€ê²½í•˜ì—¬ ì§ˆë³‘ìœ ë¬´ë¡œ ë³€í™˜

```python
df['target'] = (df['num'] > 0).astype(int)
```

```python
df.drop(columns=['num'], inplace=True)
```

```python
df.corr()[['target']].sort_values(by='target', ascending=False)
```

## ì „ì²˜ë¦¬ : ëª¨ë¸ ìž…ë ¥ ì¤€ë¹„

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["target","num"])
y = df["target"]


# ì •ê·œí™”
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ë¶„í• 
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## Tensor ë³€í™˜ > DataLoader êµ¬ì„±

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)

X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.values, dtype=torch.float32).unsqueeze(1)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
```

## ëª¨ë¸ë§

###  Multilayer Perceptron(MLP) ëª¨ë¸ ì •ì˜
ðŸ”¹ ê¸°ë³¸ êµ¬ì¡° ì„¤ê³„
- ìž…ë ¥ í”¼ì²˜ ìˆ˜: X.shape[1] 
- ì¶œë ¥: 1ê°œ (ì‹¬ìž¥ë³‘ ì—¬ë¶€ 0 or 1)
- í™œì„±í™” í•¨ìˆ˜: ReLU â†’ Sigmoid
- ì†ì‹¤ í•¨ìˆ˜: BCELoss

```python
import torch.nn as nn

class HeartDiseaseMLP(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()  # ì´ì§„ ë¶„ë¥˜ì´ë¯€ë¡œ sigmoid ì¶œë ¥
        )

    def forward(self, x):
        return self.model(x)

# ëª¨ë¸ ìƒì„±
input_dim = X_train_tensor.shape[1]
model = HeartDiseaseMLP(input_dim=input_dim)

# ìž¥ì¹˜ ì„¤ì •
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

### ì†ì‹¤í•¨ìˆ˜ ë° ì˜µí‹°ë§ˆì´ì € ì •ì˜

```python
import torch.optim as optim

criterion = nn.BCELoss()  # Binary Cross Entropy
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

### í•™ìŠµ ë£¨í”„

```python
num_epochs = 50

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    correct = 0
    total = 0

    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)

        # ìˆœì „íŒŒ
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)

        # ì—­ì „íŒŒ
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

        # ì •í™•ë„ ê³„ì‚°
        preds = (outputs > 0.5).float()
        correct += (preds == y_batch).sum().item()
        total += y_batch.size(0)

    accuracy = correct / total
    print(f"[Epoch {epoch+1}] Loss: {epoch_loss:.4f}, Accuracy: {accuracy:.4f}")
```

### ê²€ì¦(Validation)

```python
model.eval()
with torch.no_grad():
    val_preds = model(X_val_tensor.to(device))
    val_preds_binary = (val_preds > 0.5).float()
    acc = (val_preds_binary.cpu() == y_val_tensor).float().mean()
    print(f"Validation Accuracy: {acc:.4f}")
```
