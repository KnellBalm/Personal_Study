# 데이터 가져오기
- UCI 심장병 데이터 https://archive.ics.uci.edu/dataset/45/heart+disease

```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
 
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 
  
# metadata 
print(heart_disease.metadata) 
  
# variable information 
print(heart_disease.variables) 
df = pd.concat([X, y], axis=1)
    
```

## Target 역할 컬럼 정의
- 심장병 유무를 나타내는 이진 값 컬럼이 있는 지 확인
- num 컬럼은 0 이면 정상, 1~4는 질환이 있기 때문에 boolean 값으로 변경하여 질병유무로 변환

```python
df['target'] = (df['num'] > 0).astype(int)
```

```python
df.drop(columns=['num'], inplace=True)
```

```python
df.corr()[['target']].sort_values(by='target', ascending=False)
```

## 전처리 : 모델 입력 준비

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["target","num"])
y = df["target"]


# 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 분할
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## Tensor 변환 > DataLoader 구성

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)

X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.values, dtype=torch.float32).unsqueeze(1)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
```

## 모델링

###  Multilayer Perceptron(MLP) 모델 정의
🔹 기본 구조 설계
- 입력 피처 수: X.shape[1] 
- 출력: 1개 (심장병 여부 0 or 1)
- 활성화 함수: ReLU → Sigmoid
- 손실 함수: BCELoss

```python
import torch.nn as nn

class HeartDiseaseMLP(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()  # 이진 분류이므로 sigmoid 출력
        )

    def forward(self, x):
        return self.model(x)

# 모델 생성
input_dim = X_train_tensor.shape[1]
model = HeartDiseaseMLP(input_dim=input_dim)

# 장치 설정
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

### 손실함수 및 옵티마이저 정의

```python
import torch.optim as optim

criterion = nn.BCELoss()  # Binary Cross Entropy
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

### 학습 루프

```python
num_epochs = 50

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    correct = 0
    total = 0

    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)

        # 순전파
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)

        # 역전파
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

        # 정확도 계산
        preds = (outputs > 0.5).float()
        correct += (preds == y_batch).sum().item()
        total += y_batch.size(0)

    accuracy = correct / total
    print(f"[Epoch {epoch+1}] Loss: {epoch_loss:.4f}, Accuracy: {accuracy:.4f}")
```

```python
print(y_train_tensor.min(), y_train_tensor.max())
print(y_train_tensor.dtype)
```

### 검증(Validation)

```python
model.eval()
with torch.no_grad():
    val_preds = model(X_val_tensor.to(device))
    val_preds_binary = (val_preds > 0.5).float()
    acc = (val_preds_binary.cpu() == y_val_tensor).float().mean()
    print(f"Validation Accuracy: {acc:.4f}")
```
